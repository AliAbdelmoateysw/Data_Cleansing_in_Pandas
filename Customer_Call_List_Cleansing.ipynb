```python
import  pandas as  pd 
```

```python
# Read the Excel file into a DataFrame
df = pd.read_excel('Data/Customer Call List.xlsx')
df
```

```python
#Clean Last_Name
df['Last_Name']= df['Last_Name'].str.strip('/_...')
```

```python
#Clean Phone_Number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df= df.astype({'Phone_Number':str})
df['Phone_Number'] =df['Phone_Number'].apply(lambda  x: x[0:3] + '-'+x[3:6]+'-'+x[6:10] ) 
df['Phone_Number']=df['Phone_Number'].replace('nan--','').replace('Na--','')
```

```python
#Clean Address
df[['Street_Address','Country_Name','Zip']] = df['Address'].str.split(',',2,expand=True)
```

```python
#Clean Paying Customer, Do_Not_Contact
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y').replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y').replace('No','N')
```

```python
#Drop columns
columns_to_drop = ['Not_Useful_Column','Unnamed: 8','Unnamed: 9']
df= df.drop(columns_to_drop,axis=1)
df=df.replace('N/a','')
df=df.fillna('')
```

```python
#Select desired values
df = df[df['Do_Not_Contact']!='Y']
df = df[df['Phone_Number']!='']
```

```python
#Reset index and save 
df=df.reset_index(drop=True)
df.to_csv('Data/cleaned_data.csv',index=False)
```
